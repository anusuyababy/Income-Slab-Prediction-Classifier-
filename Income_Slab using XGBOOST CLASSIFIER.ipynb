SO THIS IS A CLASSIFICATION PROBLEM STATEMENT. SO WE HAVE TO CONVERT TARGET NUMERICAL VARIABLE TO CATEGORICAL VARIABLE

EDA

```python
df.income.unique()
```

```python
#converting target numerical value to categorical value
g={'<=50K' : 0, '>50K' : 1 }
df['income']=df['income'].map(g)
df.head()
```

HANDLING MISSING VALUES

```python
df=df.replace('?', np.NaN)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.workclass.unique()
```

```python
df.occupation.unique()
```

```python
df['native.country'].unique()
```

```python
df.workclass.value_counts()
```

```python
df.occupation.value_counts()
```

```python
df['native.country'].value_counts()
```

```python
df['workclass']=df['workclass'].fillna('Private')
df['occupation']=df['occupation'].fillna('Other')
df['native.country']=df['native.country'].fillna('United-States')
df.head()
```

```python
df.education.value_counts()
```

FEATURE ENGINEERING

```python
df.columns
```

```python
col = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
```

```python
df['native.country'].value_counts().to_dict()
```

```python
for feature in col:
    labels_ordered=df[feature].value_counts().to_dict()
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
df.head()
```

```python
x = df.drop(['income'], axis=1)
y = df.income
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size =0.20, random_state = 3)
```

MODEL CREATION - XGBOOST CLASSIFIER

```python
import xgboost
```

```python
params={'learning_rate': [0.05, 0.10, 0.15, 0.20, 0.25, 0.30],
        'max_depth' : [3, 4, 5, 6, 8, 10, 12, 15, 20, 25, 30],
         'min_child_weight' : [1, 3, 5, 7],
        'gamma' : [0.0, 0.1, 0.2, 0.3, 0.4], 
        'colsample_bytree' : [0.3, 0.4, 0.5, 0.7]}
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
clf=xgboost.XGBClassifier()
random_search=RandomizedSearchCV(clf, param_distributions=params, n_iter=5, scoring='roc_auc', n_jobs=-1, cv=5, verbose=3)
random_search.fit(x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_params_
```

```python
clf=xgboost.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.4, gamma=0.2,
              learning_rate=0.25, max_delta_step=0, max_depth=6,
              min_child_weight=1, missing=None, n_estimators=100, n_jobs=1,
              nthread=None, objective='binary:logistic', random_state=0,
              reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
              silent=None, subsample=1, verbosity=1)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(clf, x_train, y_train, cv=10)
```

```python
score
```

```python
score.mean()
```

```python
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_test,y_pred))
```
